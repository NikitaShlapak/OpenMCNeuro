```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def step(i, k=1.2):
    return k**i

def prepare_for_camp(k=1.2):
    steps = [1]
    i=1
    while sum(steps)<24*300:
        steps.append(step(i, k))
        i+=1
    return steps

def smooth_step(_from=1, _to=24, k=1.1):
    steps = [_from]
    while steps[-1]<_to:
        steps.append(steps[-1]*k)
    steps[-1]=_to
    return steps
```

```python
len(smooth_step())
```

```python
fig, ax = plt.subplots(figsize=(10,10))
for i in np.arange(1.05, 2, 0.05):
    camp = prepare_for_camp(i)
    ax.plot(camp, label=f'{i=:.2f} | Steps: {len(camp)} | Total: {sum(camp)//24}')
    ax.legend()
```
